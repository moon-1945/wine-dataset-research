```python
scaler = StandardScaler()

X = white_wine.drop('quality', axis=1)
X.drop('density', axis=1, inplace=True)
X['free_so2/tot_so2'] = X['free sulfur dioxide'] / X['total sulfur dioxide']
X[:] = scaler.fit_transform(X[:])
X = sm.add_constant(X)

y = white_wine['quality'] 
```

```python
def process_features_subset(y, X, feature_subset):
    X_ftr = X[list(feature_subset)]
    model = sm.MNLogit(y, X_ftr)
    regr = model.fit()
    deviance = -2 * regr.llf
    return {'model': regr, 'deviance': deviance}

def get_best_k(features_num, y, X):
    results = [process_features_subset(y, X, combo) for combo in itertools.combinations(X.columns, features_num) if 'const' in combo]
    models = pd.DataFrame(results)
    best_model = models.loc[models['deviance'].argmin()]
    return best_model

def get_best_models(y, X):
    best_models = pd.DataFrame([get_best_k(features_num, y, X) for features_num in range(1, len(X.columns) + 1)])
    best_models['pred_len'] = best_models['model'].apply(lambda m: len(m.model.exog_names))
    return best_models
```

```python
best_models = get_best_models(y, X)
```

```python
best_models['pred_len'] = best_models['model'].apply(lambda m: len(m.model.exog_names))
```

```python
def plot_models_stats(best_models_):
    plt.figure(figsize=(30,10))
    plt.rcParams.update({'font.size': 18, 'lines.markersize': 10})
    
    plt.subplot(1, 3, 1)
    
    plt.plot(best_models_['pred_len'], best_models_['deviance'])
    plt.xlabel('# Predictors')
    plt.ylabel('deviance')
    
    aic = best_models_.apply(lambda row: row['model'].aic, axis=1)
    
    plt.subplot(1, 3, 2)
    plt.plot(best_models_['pred_len'], aic)
    plt.plot(aic.argmin(), aic.min(), "or")
    plt.xlabel('# Predictors')
    plt.ylabel('AIC')
    
    bic = best_models_.apply(lambda row: row['model'].bic, axis=1)
    
    plt.subplot(1, 3, 3)
    plt.plot(best_models_['pred_len'], bic)
    plt.plot(bic.argmin(), bic.min(), "or")
    plt.xlabel('# Predictors')
    plt.ylabel('BIC')
```

```python
plot_models_stats(best_models)
```

```python
#forward selection
```

```python
def forward_step(y, X, predictors):
    remaining_predictors = [p for p in X.columns if p not in predictors]
    results = [process_features_subset(y, X, predictors+[p]) for p in remaining_predictors]
    models = pd.DataFrame(results)
    best_model = models.loc[models['deviance'].argmin()]
    return best_model

def forward(y, X):
    models_fwd = pd.DataFrame(columns=['model', 'deviance'])
    predictors = []
    for i in range(1,len(X.columns)+1):
        models_fwd.loc[i] = forward_step(y, X, predictors)
        predictors = models_fwd.loc[i]["model"].model.exog_names
    models_fwd['pred_len'] = models_fwd['model'].apply(lambda m: len(m.model.exog_names))
    return models_fwd
```

```python
forward_models = forward(y, X)
```

```python
plot_models_stats(forward_models)
```

```python
def backward_step(y, X, predictors):
    results = [process_features_subset(y, X, combo) for combo in itertools.combinations(predictors, len(predictors)-1)]
    models = pd.DataFrame(results)
    best_model = models.loc[models['deviance'].argmin()]
    return best_model

def backward(y, X):
    models_bwd = pd.DataFrame(columns=['model', 'deviance'], index = range(1,len(X.columns)))
    predictors = X.columns
    while(len(predictors) > 1):  
        models_bwd.loc[len(predictors)-1] = backward_step(y, X, predictors)
        predictors = models_bwd.loc[len(predictors)-1]["model"].model.exog_names
    models_bwd['pred_len'] = models_bwd['model'].apply(lambda m: len(m.model.exog_names))
    return models_bwd
```

```python
backward_models = backward(y, X)
```

```python
plot_models_stats(backward_models)
```

```python
train_errors = []
test_errors = []
min_test_error = 100
min_train_error = 100
best_n_neighbors = 1
n_neighbors_range = np.arange(1, 20)

for n_neighbors in n_neighbors_range:
    model_neigh = KNeighborsClassifier(n_neighbors=n_neighbors)
    model_neigh.fit(X_train, y_train)

    y_pred_train = model_neigh.predict(X_train)
    y_pred_test = model_neigh.predict(X_test)

    train_mse = mean_squared_error(y_train, y_pred_train)
    test_mse = mean_squared_error(y_test, y_pred_test)

    train_errors.append(train_mse)
    test_errors.append(test_mse)
    
    if test_mse < min_test_error and train_mse < min_train_error:
        min_test_error = test_mse
        min_train_error = train_mse
        best_n_neighbors = n_neighbors
```

```python
plt.plot(1 / n_neighbors_range, train_errors, label='train_mse')
plt.plot(1 / n_neighbors_range, test_errors, label='test_mse')
plt.xscale('log')
plt.xlabel('1/K')
plt.ylabel('mse')
plt.legend()
plt.show()
```

```python
X = white_wine.drop('quality', axis=1)
y = white_wine['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
train_errors = []
test_errors = []
depth_range = range(1,16)

for depth in depth_range:
    reg_tree = DecisionTreeRegressor(
        criterion='squared_error',
        max_depth=depth,
        min_samples_split=10,
        min_samples_leaf=5
    )

    reg_tree.fit(X_train, y_train)

    y_pred_train = reg_tree.predict(X_train)
    y_pred_test = reg_tree.predict(X_test)

    train_error = mean_squared_error(y_train, y_pred_train)
    test_error = mean_squared_error(y_test, y_pred_test)

    train_errors.append({'depth': depth, 'train_error': train_error})
    test_errors.append({'depth': depth, 'test_error': test_error})

train_errors = pd.DataFrame(train_errors)
test_errors = pd.DataFrame(test_errors)
```

```python
best_depth_idx = test_errors['test_error'].argmin()

plt.plot(train_errors['depth'], train_errors['train_error'], label='train_errors')
plt.plot(test_errors['depth'], test_errors['test_error'], label='test_errors')
plt.scatter(test_errors['depth'][best_depth_idx], test_errors['test_error'][best_depth_idx], c='r', zorder=5, label='min_train_error')
plt.legend() # test plot have high variance
```

```python
scaler = StandardScaler()

X = white_wine.drop('quality', axis=1)
y = white_wine['quality']
X[:] = scaler.fit_transform(X[:])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
class WineDataset(Dataset):
    def __init__(self, data, labels):
        self.data = torch.tensor(data, dtype=torch.float32)
        self.labels = torch.tensor(labels, dtype=torch.float32)

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx], self.labels[idx]
```

```python
train_data = WineDataset(np.array(X_train), np.array(y_train))
test_data = WineDataset(np.array(X_test), np.array(y_test))
train_data, test_data
```

```python
BATCH_SIZE = 16
NUM_WORKERS = 0

train_dataloader = DataLoader(dataset=train_data,
                              batch_size=BATCH_SIZE,
                              num_workers=NUM_WORKERS,
                              shuffle=True)

test_dataloader = DataLoader(dataset=test_data,
                              batch_size=BATCH_SIZE,
                              num_workers=NUM_WORKERS,
                              shuffle=False)

train_dataloader, test_dataloader, NUM_WORKERS
```

```python
next(iter(train_dataloader))[0].shape
```

```python
class QualityClassifier(nn.Module):
    def __init__(self, n_features):
        super(QualityClassifier, self).__init__()

        self.classifier = nn.Sequential(
            nn.Linear(n_features, 128),
            nn.ReLU(),
            nn.Linear(128, 64),
            nn.ReLU(),
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32,1)
        )

    def forward(self, x):
        return self.classifier(x)
```

```python
model = QualityClassifier(11)
```

```python
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(params=model.parameters(),
                            lr=0.001)
```

```python
from tqdm.auto import tqdm

epochs = 200

for epoch in range(epochs):

  model.train()
  train_loss = 0
  for X, y in train_dataloader:

    y_pred = model(X)

    loss = loss_fn(y_pred, y)
    train_loss += loss

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
      
  train_loss /= len(train_dataloader)

  test_loss = 0
  model.eval()
  with torch.inference_mode():
    for X_test, y_test in test_dataloader:
      test_pred = model(X_test)

      test_loss += loss_fn(test_pred, y_test)

    test_loss /= len(test_dataloader)

  
  if epoch % 10 == 0:
    print(f"\nEpoch: {epoch} | Train loss: {train_loss:.5f} | Test loss: {test_loss:.5f}\n")
```
